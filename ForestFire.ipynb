```python
import matplotlib.pylab as plt
import matplotlib as mlab
import numpy as np
%matplotlib notebook
plt.style.use('dark_background')

def setupForest(nx,ny,p):
    """
        This function sets up and plots the initial state of the forest.
    """
    forestColors = mlab.colors.ListedColormap(('white','green','red','black'))
    probOfTree=p
    forest = np.zeros((nx, ny))
    for i in range(nx):
        for j in range(ny):
            #print (i,j)
            p = np.random.rand()
            if p <= probOfTree:    # site has tree if p < probOfTree
                forest[i,j]=0.3   # color of grid site set to green
                #print (i,j)
            else:
                forest[i,j]=0.0   # unoccupied sites are white]
                #print( i,j)

    plt.pcolor(forest,cmap=forestColors,vmin=0.0,vmax=1.0)
    plt.colorbar()
    plt.show()

    return(forest)


def igniteLeftColumn(array):
    """
    This function ignites the leftmost column of the passed array.
    Only cells that have a tree are ignited.
    Input: array
    Returns: array
    """
    #need to implement this!
    return None
```
